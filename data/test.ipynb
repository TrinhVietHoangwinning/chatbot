```python
import os


data_path = 'C:\project\chatbot\data\jutsus.jsonl'

if not os.path.exists(data_path):
    raise FileNotFoundError(f"File không tồn tại: {data_path}")

with open(data_path, "r", encoding="utf-8") as f:
    lines = f.readlines()
    print("Số dòng trong file:", len(lines))
    print("Dòng đầu tiên:", lines[:2])  # In thử 2 dòng đầu
```

```python
import json

with open(data_path, "r", encoding="utf-8") as f:
    for i, line in enumerate(f):
        try:
            json.loads(line)  # Kiểm tra từng dòng
        except json.JSONDecodeError as e:
            print(f"Lỗi JSON tại dòng {i+1}: {e}")
```

```python
import pandas as pd

df = pd.read_json(data_path, lines=True, encoding="utf-8")
print(df.head())  # Xem thử 5 dòng đầu tiên
```
